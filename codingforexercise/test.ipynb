```python
def rc4_decrypt(key, ciphertext):
    # Key Scheduling Algorithm (KSA)
    S = list(range(256))
    j = 0
    for i in range(256):
        j = (j + S[i] + key[i % len(key)]) % 256
        S[i], S[j] = S[j], S[i]

    # Pseudo-Random Generation Algorithm (PRGA)
    i = 0
    j = 0
    plaintext = []
    for byte in ciphertext:
        i = (i + 1) % 256
        j = (j + S[i]) % 256
        S[i], S[j] = S[j], S[i]
        k = S[(S[i] + S[j]) % 256]
        plaintext.append(byte ^ k)

    return bytes(plaintext)

def main():
    hex_string = "A71A68ECD82711CC8C9B16155CD2673E82ADCE75D4BC5756C28A52B86BD6CCF8A4BA722FE05715B92411"
    key_string = "RC4_1s_4w3s0m3"
    
    # å°†åå…­è¿›åˆ¶å­—ç¬¦ä¸²è½¬æ¢ä¸ºå­—èŠ‚
    ciphertext = bytes.fromhex(hex_string)
    # å°†å¯†é’¥å­—ç¬¦ä¸²è½¬æ¢ä¸ºå­—èŠ‚
    key = key_string.encode()
    
    # ä½¿ç”¨ RC4 è§£å¯†
    decrypted_data = rc4_decrypt(key, ciphertext)
    
    # å°è¯•è§£ç ä¸º UTF-8 å­—ç¬¦ä¸²
    try:
        plaintext = decrypted_data.decode('utf-8')
        print("è§£å¯†åçš„æ˜æ–‡:", plaintext)
    except UnicodeDecodeError:
        print("è§£å¯†ç»“æœä¸æ˜¯æœ‰æ•ˆçš„ UTF-8 å­—ç¬¦ä¸²ï¼Œä»¥åå…­è¿›åˆ¶æ˜¾ç¤º:")
        print(decrypted_data.hex())

if __name__ == "__main__":
    main()
```

```python
import json
from Crypto.Util.number import long_to_bytes, bytes_to_long
from math import gcd

# --- LCG å’Œ mix/unmix å‡½æ•° (æœ€ç»ˆä¿®æ­£ç‰ˆ) ---
A = 1103515245
B = 12345
MOD = 2**31

def lcg(s):
    """çº¿æ€§åŒä½™ç”Ÿæˆå™¨"""
    while True:
        s = (A * s + B) % MOD
        yield s & 0xFF

def unmix(b, s):
    """
    ã€å…³é”®çš„æ­£ç¡®è§£å¯†å‡½æ•°ã€‘
    è¿™ä¸ªå‡½æ•°æ˜¯ mix å‡½æ•°çš„ç²¾ç¡®æ•°å­¦é€†ã€‚
    """
    g = lcg(s)
    o = bytearray()
    
    # æŒ‰8å­—èŠ‚å—å¤„ç†åŠ å¯†æ•°æ®
    for i in range(0, len(b), 8):
        chunk = b[i:i+8]
        
        # åˆ›å»ºä¸€ä¸ªä¸´æ—¶çš„ã€å­—èŠ‚é¡ºåºä»ç„¶æ˜¯é¢ å€’çš„è§£å¯†å—
        temp_decrypted_chunk = bytearray()
        for byte_val in chunk:
            temp_decrypted_chunk.append(byte_val ^ next(g))
            
        # å°†è¿™ä¸ªé¢ å€’çš„å—å†é¢ å€’å›æ¥ï¼Œæ¢å¤åŸå§‹é¡ºåº
        original_order_chunk = temp_decrypted_chunk[::-1]
        o.extend(original_order_chunk)
        
    return bytes(o)

# --- 1. åŠ è½½æŒ‘æˆ˜æ•°æ® ---
print("[*] Loading data from challenge.json...")
with open("challenge.json", "r") as f:
    data = json.load(f)

n = int(data["n"], 16)
e = int(data["e"])
sig_ok_mixed = bytes.fromhex(data["sig_ok_mixed"])
sig_fault_mixed = bytes.fromhex(data["sig_fault_mixed"])
flag_enc = int(data["flag_enc"], 16)
seed_hint = int(data["seed_hint"])
print("   [+] Data loaded successfully.")

# --- 2. æš´åŠ›ç ´è§£ seed ---
print("\n[*] Starting brute-force with the correct logic (this will be fast)...")
found_seed = -1
p = -1
max_k = (2**31 - 1 - seed_hint) // 97

for k in range(max_k + 1):
    seed_candidate = 97 * k + seed_hint
    
    # ä½¿ç”¨æ­£ç¡®çš„ unmix å‡½æ•°è§£å¯†
    sig_ok_bytes = unmix(sig_ok_mixed, seed_candidate)
    sig_fault_bytes = unmix(sig_fault_mixed, seed_candidate + 123456)
    
    # è½¬æ¢ä¸ºæ•´æ•°
    sig_ok_candidate = bytes_to_long(sig_ok_bytes)
    sig_fault_candidate = bytes_to_long(sig_fault_bytes)
    
    diff = abs(sig_ok_candidate - sig_fault_candidate)
    if diff == 0:
        continue
        
    p_candidate = gcd(diff, n)
    
    if 1 < p_candidate < n:
        print(f"\n   [+] Success! Found a non-trivial factor of n.")
        p = p_candidate
        found_seed = seed_candidate
        break

if found_seed == -1:
    print("\n[-] Brute-force failed. This shouldn't happen with the correct logic.")
    exit()

print(f"   [+] Correct seed found: {found_seed}")
print(f"   [+] Found prime factor p: {p}")

# --- 3. æ¢å¤ m ---
print("\n[*] Recovering the message 'm'...")
sig_ok_bytes = unmix(sig_ok_mixed, found_seed)
sig_ok = bytes_to_long(sig_ok_bytes)
m = pow(sig_ok, e, n)
print(f"   [+] Recovered m.")

# --- 4. è§£å¯† Flag ---
print("\n[*] Decrypting the flag...")
flag_long = flag_enc ^ m
flag_bytes = long_to_bytes(flag_long)

# --- æœ€ç»ˆç»“æœ ---
print("\n" + "="*40)
print(f"  DECRYPTION SUCCESSFUL!")
print(f"  FLAG: {flag_bytes.decode()}")
print("="*40)
```
